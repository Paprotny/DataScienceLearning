```python
import pandas as pd
import numpy as np
import re
import sys
```

```python
raw_data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/fJKLoMKtgpDAl5MkA6rf2A/jfk-weather.csv", parse_dates=["DATE"])
```

```python
pd.set_option('display.max_rows', 120)
pd.set_option('display.max_columns', 120)
raw_data.dtypes
```

```python
columns=[
    'DATE',
    'HOURLYVISIBILITY',
    'HOURLYDRYBULBTEMPF',
    'HOURLYWETBULBTEMPF',
    'HOURLYDewPointTempF',
    'HOURLYRelativeHumidity',
    'HOURLYWindSpeed',
    'HOURLYWindDirection',
    'HOURLYStationPressure',
    'HOURLYPressureTendency',
    'HOURLYSeaLevelPressure',
    'HOURLYPrecip',
    'HOURLYAltimeterSetting'
]
```

```python
filtered_data = raw_data[columns]
```

```python
filtered_data=filtered_data.set_index(filtered_data['DATE'], inplace=False)
filtered_data = filtered_data.drop(columns=['DATE'])
```

```python
filtered_data['HOURLYPrecip'].unique()
```

```python
filtered_data['HOURLYPrecip'] = filtered_data['HOURLYPrecip'].str.extract(r'(\d+\.\d{2})').astype(float)
```

```python
filtered_data.head()
```

```python
filtered_data['HOURLYPrecip'].unique()
```

```python
messy_columns = columns[1:]
```

```python
for i in messy_columns:
    filtered_data[i]=filtered_data[i].apply(
    lambda x: re.sub('[^0-9,.-]', '', x) if type(x) == str else x)
    filtered_data[i]=filtered_data[i].replace('', np.nan).astype(('float32'))
    
```

```python
filtered_data.describe()
```

```python
filtered_data[(filtered_data['HOURLYVISIBILITY'] < 0) | (filtered_data['HOURLYVISIBILITY'] > 10)]
```

```python
filtered_data.loc[filtered_data['HOURLYVISIBILITY'] > 10, 'HOURLYVISIBILITY'] = np.nan
```

```python
sum(filtered_data.index.duplicated())
```

```python
%%capture
filtered_data['HOURLYPressureTendency'] = filtered_data['HOURLYPressureTendency'].fillna(method='ffill') # fill with last valid observation
filtered_data = filtered_data.interpolate(method='linear') # interpolate missing values
filtered_data.drop(filtered_data.index[0], inplace=True) # drop first row
```
