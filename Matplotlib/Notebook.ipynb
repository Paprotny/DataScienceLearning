```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sys
import os
```

```python
x_1=np.linspace(0,10,100)
y_1=x_1**2
```

```python
plt.plot(x_1, y_1, label='y = x^2')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, color='blue')
plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, color='red')
```

```python
fig_1=plt.figure(figsize=(5, 4),dpi=100)
axes_1=fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title('Days vs Days squared')
axes_1.plot(x_1, y_1, color='blue', label='x/x2')
axes_1.plot(y_1, x_1, color='green', label='x2/x')
axes_1.legend(loc=0)

```

```python
fig_1=plt.figure(figsize=(5, 4),dpi=100)
axes_1=fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title('Days vs Days squared')
axes_1.plot(x_1, y_1, color='blue', label='x/x2')
axes_1.plot(y_1, x_1, color='green', label='x2/x')
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days squared')
axes_2.set_title('Days vs Days squared')
axes_2.plot(x_1,y_1,'r')
```

```python
fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1,ncols=3)
plt.tight_layout()
axes_2[1].set_title('Second plot')
axes_2[1].set_xlabel('x')
axes_2[1].set_ylabel('x squared')
axes_2[1].plot(x_1, y_1, color='blue', label='x/x2')
axes_2[1].legend(loc=0)
```

```python
fig_3=plt.figure(figsize=(6,4))
axes_3=fig_3.add_axes([0,0,1,1])
axes_3.plot(x_1,y_1, color='navy',alpha=0.75, lw=2, ls='-.', marker='o', markersize=7, markerfacecolor='y', markeredgecolor='y',markeredgewidth=4)
axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])
axes_3.grid(True, color='0.6',dashes=(5,2,1))
axes_3.set_facecolor('lightgrey')
```

```python
fig_4=plt.figure(figsize=(10,5))
axes_1=fig_4.add_axes([0.5, 0.5, 0.9, 0.9])
x_1=np.linspace(-5,5,100)
y_1=np.sin(x_1)
axes_1.set_xlabel('x')
axes_1.set_ylabel('sin(x)')
axes_1.set_title('Sine Function')
axes_1.plot(x_1,y_1, color='r', lw=2, marker='x')
axes_1.set_xlim([-np.pi, np.pi])
```

```python
fig_4.savefig('3rd_plot.png', bbox_inches='tight')
```
