### Roman to Integer

```python
def roman(string):
    dictonary={
            "I":1,
            "V":5,
            "X":10,
            "L":50,
            "C":100,
            "D":500,
            "M":1000
        }
    result=0
    counter=0
    checker=1000
    indices=[]
    for a in string:
        if dictonary[a]>checker:
            result+=dictonary[a]-checker
            indices+=[counter-1,counter]
        counter+=1
        checker=dictonary[a]
    string = [x for i, x in enumerate(string) if i not in indices]
    for i in string:
        result+=dictonary[i]
    return result
```

```python
string='MMMCDVIII'
roman(string)
```

### Roman to Integer attempt 2

```python
def romanToInt(s: str) -> int:
        numbers = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
        res = 0
        for i in range(len(s)):
            if i+1 < len(s) and numbers[s[i]] < numbers[s[i+1]]:
                res -= numbers[s[i]]
            else:
                res += numbers[s[i]]
        return res
```

```python
s='MMMCDVIII'
```

```python
romanToInt(s)
```

```python
import time

start = time.perf_counter()
romanToInt(s)
end = time.perf_counter()

print("Time performed:", end - start, "sec")
```

```python
def longestCommonPrefix(strs):
    result=min(strs,key=len)
    min_len = min(len(s) for s in strs)
    for i in strs:
        for index,data in enumerate(i):
            if index>=len(result) or data!=result[index]:
                result=result[:index]
                break
        if not result:
            break
    return result
```

```python
#Clever result
def longestCommonPrefix1(strs):
    if not strs:
            return ""
    prefix=strs[0]
    for i in strs[1:]:
        while not i.startswith(prefix):
            prefix=prefix[:-1]
            if not prefix:
                return ""
    return prefix
```

```python
class ListNode:
     def __init__(self, val=0, next=None):
         self.val = val
         self.next = next
```

```python
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        # dummy node to simplify edge cases
        dummy = ListNode(-1)
        current = dummy
        
        # iterate while both lists have nodes
        while l1 and l2:
            if l1.val < l2.val:
                current.next = l1
                l1 = l1.next
            else:
                current.next = l2
                l2 = l2.next
            current = current.next
        
        # attach the remaining nodes (if any)
        if l1:
            current.next = l1
        if l2:
            current.next = l2
        
        # return the head of the merged list
        return dummy.next
```
